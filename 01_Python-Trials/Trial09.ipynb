```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import style
import scipy
from scipy import signal
```

```python
x = [] # With this definition remember that a space in the memory is reserve to generate an array.
n = []                  # t = nT. Since we are performing uniform sampling, therefore, T=1.
for i in range(-5,6):
    n.append(i) #It means a discrete distribution of points
    #if i==0:
    #if i==2:    #Time shifted delay unit impulse signal
    if i==-2:    #Time shifted advanced unit impulse signal
        x.append(1)
    else:
        x.append(0)
print(x)
print(n)
```

```python
plt.figure(figsize = (18,8)) # set the size of figure
#plt.suptitle('Constructing a Wave with different Sine Waves', fontsize = 40)
style.use('dark_background')
plt.rcParams['xtick.labelsize'] = 15
plt.rcParams['ytick.labelsize'] = 15
plt.stem(n,x,'yo', label = 'Unit impulse function') #stem function will plot a discrete function instead of plt.plot 
#for continuous functions. The color of the points yo, y is yellow and "o" is a filled circle
plt.ylim(-0.01,1.2)
plt.xlim(-5,5)
plt.xlabel('time.', fontsize = 15)
plt.ylabel('Amplitude', fontsize = 15)
plt.legend(fontsize = 20)
plt.show()
```

```python
# Generation of time step impulse
x = [] # With this definition remember that a space in the memory is reserve to generate an array.
n = []                  # t = nT. Since we are performing uniform sampling, therefore, T=1.
for i in range(-5,10):
    n.append(i) #It means a discrete distribution of points
    #if i>=0:
    #if i>=2:    #Time shifted delay step impulse signal
    if i>=-2:    #Time shifted advanced unit impulse signal
        x.append(1)
    else:
        x.append(0)
print(x)
print(n)
```

```python
plt.figure(figsize = (18,8)) # set the size of figure
#plt.suptitle('Constructing a Wave with different Sine Waves', fontsize = 40)
style.use('dark_background')
plt.rcParams['xtick.labelsize'] = 15
plt.rcParams['ytick.labelsize'] = 15
plt.stem(n,x,'yo', label = 'Step impulse function') #stem function will plot a discrete function instead of plt.plot 
#for continuous functions. The color of the points yo, y is yellow and "o" is a filled circle
plt.ylim(-0.01,1.2)
plt.xlim(-5,9)
plt.xlabel('time.', fontsize = 15)
plt.ylabel('Amplitude', fontsize = 15)
plt.legend(fontsize = 20)
plt.show()
```

```python
#Generation ramp signal
x = []
n = []
for i in range(0,10):
    n.append(i)
    x.append(i)
print (x)
print (n)
```

```python
plt.figure(figsize = (18,8)) # set the size of figure
style.use('dark_background')
plt.rcParams['xtick.labelsize'] = 15
plt.rcParams['ytick.labelsize'] = 15
plt.stem(n,x,'yo', label = 'Ramp function')
plt.ylim(0,len(x)) #This definition of limits is used to automaticly adapt the plot to the function values.
plt.xlim(0,len(n)-1)
plt.xlabel('time.', fontsize = 15)
plt.ylabel('Amplitude', fontsize = 15)
plt.legend(fontsize = 20)
plt.show()
```

```python
#Generation of sine impuse signal
n = np.linspace(0,10,100) #Remember that this function divides an interval from 0 to 10 in a 100 time steps
amp = 5 # amplitude
f = 50  # frequency
x = amp * np.sin(2 * np.pi * f * n)
```

```python
plt.figure(figsize = (18,12)) # set the size of figure
style.use('dark_background')
plt.rcParams['xtick.labelsize'] = 15
plt.rcParams['ytick.labelsize'] = 15
plt.stem(n,x,'yo', label = 'SineWave') #Remind that this operator graphs a discrete function.
plt.ylim(-5,5)
plt.xlim(0,10)
plt.xlabel('time.', fontsize = 15)
plt.ylabel('Amplitude', fontsize = 15)
plt.legend(fontsize = 15)
plt.show()
```

```python
#Generation of an exponential impulse signal
n = np.linspace(0,10,100)
amp = 1
x = amp * np.exp(-n)
```

```python
plt.figure(figsize = (15,8)) # set the size of figure
style.use('dark_background')
plt.rcParams['xtick.labelsize'] = 15
plt.rcParams['ytick.labelsize'] = 15
plt.stem(n,x,'yo', label = 'Exponential Signal')
plt.xlabel('time.', fontsize = 15)
plt.ylabel('Amplitude', fontsize = 15)
plt.xlim(0,5)
plt.ylim(0,1)
plt.legend(fontsize = 20)
plt.show()
```
