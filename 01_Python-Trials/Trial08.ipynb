```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import style
import scipy #Scientific python
from scipy import signal
```

```python
#Generating unit impulse signal
x = []
t = []
for i in range(-5,6):
    t.append(i)
    if i==0:
        x.append(1)
    else:
        x.append(0)
#print(x)
#print(t)
```

```python
#plotting the unit ipulse signal
plt.figure(figsize = (18,8)) # set the size of figure
#plt.suptitle('Constructing a Wave with different Sine Waves', fontsize = 40)
style.use('dark_background')
plt.rcParams['xtick.labelsize'] = 15
plt.rcParams['ytick.labelsize'] = 15
plt.plot(t,x,linewidth = 3, label = 'Unit impulse function')
plt.ylim(-0.01,1)
plt.xlim(-5,5)
plt.xlabel('time.', fontsize = 15)
plt.ylabel('Amplitude', fontsize = 15)
plt.legend(fontsize = 20)
plt.show()
```

```python
#Generating the unit impulse signal using a subroutine
w = signal.unit_impulse(10,'mid') #unit_impulse is defined with two indices. The first one indicates the number of steps of the independent variable, and the second one
#indicates where the signal has a value one. In this case is in the mid range
s = np.arange(-5, 5)
print (s)
print (w)
```

```python
#Plotting the unit impulse signal
plt.figure(figsize = (18,8)) # set the size of figure
style.use('dark_background')
plt.rcParams['xtick.labelsize'] = 15
plt.rcParams['ytick.labelsize'] = 15
plt.plot(s, w, linewidth = 3, label = 'Unit impulse function')
plt.ylim(-0.01,1)
plt.xlabel('time.', fontsize = 15)
plt.ylabel('Amplitude', fontsize = 15)
plt.legend(fontsize = 20)
plt.show()
```

```python
#Time shifted signal (delay)
shifted_impulse = signal.unit_impulse(7, 2)
print(shifted_impulse)
```

```python
plt.figure(figsize = (18,8)) # set the size of figure
style.use('dark_background')
plt.rcParams['xtick.labelsize'] = 15
plt.rcParams['ytick.labelsize'] = 15
plt.plot(shifted_impulse, linewidth = 3, label = 'Shifted impulse function (Time delay)')
plt.xlabel('time.', fontsize = 15)
plt.ylabel('Amplitude', fontsize = 15)
plt.legend(fontsize = 20)
plt.show()
```

```python
#Generating time shifted advance 
x1 = []
t = []
for i in range(-5,6):
    t.append(i)
    if i==-2:
        x1.append(1)
    else:
        x1.append(0)
print(x1)
print(t)
```

```python
#Plotting the time shifted signal (advance)
plt.figure(figsize = (18,8)) # set the size of figure
#plt.suptitle('Constructing a Wave with different Sine Waves', fontsize = 40)
style.use('dark_background')
plt.rcParams['xtick.labelsize'] = 15
plt.rcParams['ytick.labelsize'] = 15
plt.plot(t,x1,linewidth = 3, label = 'Unit impulse function')
plt.ylim(-0.01,1)
plt.xlim(-5,5)
plt.xlabel('time.', fontsize = 15)
plt.ylabel('Amplitude', fontsize = 15)
plt.legend(fontsize = 20)
plt.show()
```

```python
#Generation of unit step signal
x2 = []
t = []
for i in range(-5,6):
    t.append(i)
    if i>=0:
        x2.append(1)
    else:
        x2.append(0)
print(x2)
print(t)
```

```python
#plotting the one step signal
plt.figure(figsize = (18,8)) # set the size of figure
#plt.suptitle('Constructing a Wave with different Sine Waves', fontsize = 40)
style.use('dark_background')
plt.rcParams['xtick.labelsize'] = 15
plt.rcParams['ytick.labelsize'] = 15
plt.plot(t,x2,linewidth = 3, label = 'Unit impulse function')
plt.ylim(-0.01,1.1)
plt.xlim(-5,5)
plt.xlabel('time.', fontsize = 15)
plt.ylabel('Amplitude', fontsize = 15)
plt.legend(fontsize = 20)
plt.show()
```

```python
#time shifted one step signal (advance)
x3 = []
t = []
for i in range(-5,6):
    t.append(i)
    if i>=-2:
        x3.append(1)
    else:
        x3.append(0)
print(x3)
print(t)
```

```python
#plotting the time shifted one step signal (advance)
plt.figure(figsize = (18,8)) # set the size of figure
#plt.suptitle('Constructing a Wave with different Sine Waves', fontsize = 40)
style.use('dark_background')
plt.rcParams['xtick.labelsize'] = 15
plt.rcParams['ytick.labelsize'] = 15
plt.plot(t,x3,linewidth = 3, label = 'Unit impulse function')
plt.ylim(-0.01,1.1)
plt.xlim(-5,5)
plt.xlabel('time.', fontsize = 15)
plt.ylabel('Amplitude', fontsize = 15)
plt.legend(fontsize = 20)
plt.show()
```

```python
#Generating ramp signal
x4 = []
t = []
for i in range(0,10):
    t.append(i)
    x4.append(i)
print(x4)
print(t)
```

```python
#Plotting the ramp signal
plt.figure(figsize = (18,8)) # set the size of figure
#plt.suptitle('Constructing a Wave with different Sine Waves', fontsize = 40)
style.use('dark_background')
plt.rcParams['xtick.labelsize'] = 15
plt.rcParams['ytick.labelsize'] = 15
plt.plot(t,x4,linewidth = 3, label = 'Ramp impulse signal')
plt.ylim(-0.01,10.1)
plt.xlim(0,10)
plt.xlabel('time.', fontsize = 15)
plt.ylabel('Amplitude', fontsize = 15)
plt.legend(fontsize = 20)
plt.show()
```

```python
#Generating a cosine or sine function
t = np.linspace(0,10,100)
amp = 5 # amplitude
f = 50  # frequency
x5 = amp * np.sin(2 * np.pi * f * t)
x6 = amp * np.cos(2 * np.pi * f * t)
```

```python
#plotting the sine or cosine signal
plt.figure(figsize = (18,8)) # set the size of figure
#plt.suptitle('Constructing a Wave with different Sine Waves', fontsize = 40)
style.use('dark_background')
plt.rcParams['xtick.labelsize'] = 15
plt.rcParams['ytick.labelsize'] = 15
plt.plot(t,x6,linewidth = 3, label = 'Cosine impulse function')
plt.plot(t,x5,'r',linewidth = 3, label = 'Sine impulse function')
plt.ylim(-5.01,6.01)
plt.xlim(0,10)
plt.xlabel('time.', fontsize = 15)
plt.ylabel('Amplitude', fontsize = 15)
plt.legend(fontsize = 20)
plt.show()
```

```python
#Generating Exponential Signal
t = np.linspace(0,10,100)
amp = 1 # amplitude
x7 = amp * np.exp(-t)
#print(t)
#print(x7)
```

```python
#Plotting exponential signal
plt.figure(figsize = (18,8)) # set the size of figure
#plt.suptitle('Constructing a Wave with different Sine Waves', fontsize = 40)
style.use('dark_background')
plt.rcParams['xtick.labelsize'] = 15
plt.rcParams['ytick.labelsize'] = 15
plt.plot(t,x7,linewidth = 3, label = 'Exponential impulse signal')
plt.ylim(-0.01,1.01)
plt.xlim(0,10)
plt.xlabel('time.', fontsize = 15)
plt.ylabel('Amplitude', fontsize = 15)
plt.legend(fontsize = 20)
plt.show()
```
