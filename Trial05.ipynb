```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import style

print (plt.style.available)

t = np.linspace(0,10,100)
A = 5 #Amplitude
f = 50 #Frequency
#A sine wave
x = A * np.sin(2 * np.pi * f * t)
style.use('dark_background')
print (t)
print (x)

plt.figure(figsize = (16,8)) #Changing the size of the plot
#To put a label on the plot
#plt.plot (t,x, 'r', linewidth = 3, label = 'Sine Wave')  #How to change the color of the line and linewidth (2 is the default)
#Fancier plot connects lines with dash lines and mark the points with circles/figures that have center and edge in
#different colors.
plt.plot (t,x, 'r', ls = '--',lw = 2, marker = 's', markersize = 10, markerfacecolor = 'g', markeredgecolor = 'y', label = 'Sine Wave')  #How to change the color of the line and linewidth (2 is the default)
#To label the axes of just the last plot
plt.xlabel ('Time', fontsize = 15)
plt.ylabel ('Amplitude', fontsize = 15)
plt.xlim (0,10) #Setting the limits for the x-axes
plt.ylim (-8,8) #Setting the limit for the y-axes
#plt.xticks ([0,10]) #Removes marks in the x-axes
#plt.yticks ([-8,8]) #Removes marks in the y-axes 
#plt.axis('off') #Removes axes from the plot
plt.grid() #Add a grid to the plot
plt.legend (fontsize = 15) #This command is needed for the size of legends on the plot
#If you want to include a plot title
#plt.title('Bannanas', fontsize = 25) #I do not like to use titles
plt.show()

plt.subplots(2,2) #Produces a matrix with 2 by 2 plots
plt.tight_layout() #Fit them tightly so they do not overlap
```
